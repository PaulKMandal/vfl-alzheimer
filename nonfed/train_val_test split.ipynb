```python
import pandas as pd
```

```python
df = pd.read_csv('../examples/data/oasis2/oasis2.csv')
```

```python
df
```

```python
id_list = df['id'].tolist()
```

```python
subject_id = [s[:9] for s in id_list]
```

```python
df['Subject ID'] = subject_id
```

```python
df
```

```python
unique_ids = df['Subject ID'].unique()
```

```python
train_ids = unique_ids[0:102]
val_ids = unique_ids[102:122]
test_ids = unique_ids[122:142]
```

```python
df_train = df[df['Subject ID'].isin(train_ids)]
```

```python
df_val = df[df['Subject ID'].isin(val_ids)]
```

```python
df_test = df[df['Subject ID'].isin(test_ids)]
```

```python
import shutil
import os
```

```python
source_dir = '../examples/data/oasis2/mri/img/'

target_dir = '../examples/data/oasis2train/mri/img/'

os.makedirs(target_dir, exist_ok=True)  

for i in df_train['id']:
    shutil.copy(source_dir+i+'.jpg', target_dir)

df_train = df_train.drop('Subject ID', axis=1)

df_train.to_csv(target_dir + '../../oasis2.csv', index=False)
```

```python
source_dir = '../examples/data/oasis2/mri/img/'

target_dir = '../examples/data/oasis2val/mri/img/'

os.makedirs(target_dir, exist_ok=True)  

for i in df_val['id']:
    shutil.copy(source_dir+i+'.jpg', target_dir)
    
df_val = df_val.drop('Subject ID', axis=1)

df_val.to_csv(target_dir + '../../oasis2.csv', index=False)
```

```python
source_dir = '../examples/data/oasis2/mri/img/'

target_dir = '../examples/data/oasis2test/mri/img/'

os.makedirs(target_dir, exist_ok=True)  

for i in df_test['id']:
    shutil.copy(source_dir+i+'.jpg', target_dir)
    
df_test = df_test.drop('Subject ID', axis=1)

df_test.to_csv(target_dir + '../../oasis2.csv', index=False)
```
